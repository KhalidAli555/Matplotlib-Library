```python
import matplotlib.pyplot as plt
```

 # **Linear Graph**

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y,c='r')   #change the color of line
plt.show()
```

```python
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.plot(x,y)
plt.show()
```

```python
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.plot(x,y,marker='o')  # marker make dot point in graph
plt.show()
```

```python
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.plot(x,y)
plt.plot(y,marker='>')
plt.plot(x,marker='o')  # marker make dot point in graph
plt.show()
```

# **Bar Graph**

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.bar(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
c=['r','g','b','y']  # give different color to every one
plt.bar(x,y,color=c)
plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.bar(x,y,color="green")   # Give same color to every one
plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.barh(x,y)   # Make horizontal  bar graph
plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.bar(x,y, width=0.3)   # change the width of the graph

plt.show()
```

# **Scatter Plot**

```python
x=[1,2,3,4,5,6]
y=[4,2,3,5,1,8]
plt.title("Scatter Plot")   # give title to scatter plot
plt.xlabel("Months")  # x axis label
plt.ylabel("Numbers")  # y axis label
plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,2,3,5,1,8]
plt.title("Scatter Plot")   # give title to scatter plot
plt.xlabel("Months")  # x axis label
plt.ylabel("Numbers")  # y axis label
plt.scatter(x,y ,color='r')  # give  same color to  all points
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,2,3,5,1,8]
c=["red","green","black","yellow","orange","blue"]
plt.title("Scatter Plot")   # give title to scatter plot
plt.xlabel("Months")  # x axis label
plt.ylabel("Numbers")  # y axis label
plt.scatter(x,y ,color=c ,s=100)  # give different color to every point $ 's' change the size of bubble
plt.show()
```

# **Pi Chart**

```python
x=[50,20,30,40,70]
y=["English","Urdu","Computer","Physics","Data Science"]
plt.pie(x,labels=y)
plt.show()
```

```python
x=[50,20,30,40,70]
y=["English","Urdu","Computer","Physics","Data Science"]
plt.pie(x,labels=y)
plt.legend()  # give list to all subject in one side
plt.show()
```
